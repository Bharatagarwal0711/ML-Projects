```python
# --------------------------------------------------------------------- Functions of Special Checks ----------------------------------------------------------------------------

def Repeated_Sequence(pwd:str)->bool:  # aaaa1111
    
    count = 1
    for i in range(1,len(pwd)):
        if(pwd[i]==pwd[i-1]): 
            count += 1
            if count >= 4: return True
        else: 
            count = 1

    return False


def Sequence(pwd:str)->bool:   # abcd1234

    Patterns = ['qwertyuiop','asdfghjkl','zxcvbnm','1234567890','abcdefghijklmnopqrstuvwxyz']
    size = 4
    pwd = pwd.lower()

    for row in Patterns:

        for i in range(len(row) - size+1):  # Forward Pattern
            pattern = row[i:i+size]
            if pattern in pwd: return True

        row_rev = row[::-1]
        for i in range(len(row_rev) - size+1):  # Reverse Pattern
            pattern = row_rev[i:i+size]
            if pattern in pwd: return True

    return False


import math
def Entropy_Score(n:int,pool_size:int)->int:
    
    if(pool_size == 0): return 0

    value = n * math.log2(pool_size)

    if value >= 78.6: return 5
    elif value >= 62.88: return 4
    elif value >= 47.16: return 3
    elif value >= 31.44: return 2
    elif value >= 15.72: return 1
    else: return 0


 #--------------------------------------------------------------------------- Main Function for Password Strength ---------------------------------------------------------------

def Password_Strength(pwd):
    up = False
    low = False
    digit = False
    special = False


    #----------------------------------------------------------------------------- Checkind Password Length --------------------------------------------------------------------

    if(len(pwd)<12):
        print("Password should contains atleast 12 characters") 
        print("Try Again")
        return
    

    #--------------------------------------------------------------------------------- Common Passwords ------------------------------------------------------------------------
    
    common_passwords = set([
    "121212","112233","123123","password","password123","admin","admin123","welcome","welcome123","login","guest","root","user","test","default","access","qwerty","qwerty123","asdfgh","zxcvbn","qazwsx","1q2w3e4r","abc123","letmein","iloveyou","monkey","dragon","football","baseball","master","hello123","india123","pass123","password1","welcome1","admin1","superman","batman","trustno1","freedom","secret","shadow","princess","sunshine","flower","love123","computer","internet","google","microsoft","apple123","windows","android","linkedin","facebook","instagram"
    ])  # Data From Google
    
    if pwd.lower() in common_passwords:
        print(f"Very Common Password !")
        print("Try Something else")
        return

    #----------------------------------------------------------------------------------------- Special Checks ------------------------------------------------------------------

    if Sequence(pwd) or Repeated_Sequence(pwd) : 
        print("Sequence is Not Allowed in Password ")
        print("Try Again")
        return 

    # ----------------------------------------------------------------------------------- Checking Password Strength -------------------------------------------------------------

    for i in pwd:
        if i.isupper(): up = True
        elif i.islower(): low = True
        elif i.isdigit(): digit = True
        else: special = True 

    count = 0
    pool = 0
    diversity = len(set(pwd)) / len(pwd)

    if up: 
        count += 1
        pool += 26
    if low: 
        count += 1
        pool += 26

    if digit: 
        count += 1
        pool += 10

    if special : 
        count += 1
        pool += 32

    if diversity >= 0.7: count += 1

    count += Entropy_Score(len(pwd),pool)
    count //= 2

    match count:
        case 0: print("very Weak")
        case 1: print("weak")
        case 2: print("Decent")
        case 3: print("Good")
        case 4: print("Very Good")
        case _: print("Strong")



Password_Strength("$1a@hsjvdbhsj")
```
